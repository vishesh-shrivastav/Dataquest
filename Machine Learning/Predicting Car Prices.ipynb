```python
# Fix the dataframe
colnames_to_get = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv("imports-85.data", names = colnames_to_get)
```

```python
cars.head()
```

**Data Cleaning**

1. Replace '?' with NaN

```python
cars = cars.replace("?", np.nan)
```

Drop non-numeric columns

```python
cols_to_drop = ["symboling", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels", "engine-location", 
                "engine-type", "num-of-cylinders", "fuel-system", "engine-size"]
cars_final = cars.drop(cols_to_drop, axis=1)
```

```python
cars_final.head()
```

```python
cars_final = cars_final.astype("float")
cars_final.isnull().sum()
```

We are going to use the knn algorithm to determine the price of a car. It doesn't make sense to keep rows with missing values in the 'price' column. So we'll just drop them entirely.

For the 'bore' and 'stroke' columns, we'll use the mean to fill in the missing values.

```python
cars_final = cars_final.dropna(subset=["price"])
cars_final.isnull().sum()
```

```python
# Replace missing values in other columns using column means.
cars_final = cars_final.fillna(cars_final.mean())
# Check that there are no more missing values.
cars_final.isnull().sum()
```
